```python
file2='DataForFigure2.1WHR2021C2.xls'
df2=pd.read_excel(file2)
df2 = df2[['Country name','Regional indicator']]
df2.head()
```

```python

df=pd.merge(df, df2,on='Country name')
df=df.drop(['year'], axis=1)
#print(df.head())
df.describe()
```

```python
import matplotlib.pyplot as plt
corr=df.corr()
#plt.matshow(df.corr())
#plt.show()

corr.style.background_gradient(cmap='coolwarm')
```
